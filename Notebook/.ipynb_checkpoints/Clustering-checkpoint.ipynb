```python
# Tous les imports nécessaires

import datetime
import math
import os
import sqlite3

import joblib
import keras
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
from keras.layers import GRU, LSTM, Dense, Input, RepeatVector, TimeDistributed
from keras.models import Model, load_model
from scipy import interpolate
from sklearn import preprocessing
from sklearn.cluster import Birch, KMeans
from tensorflow.keras.utils import plot_model
from tqdm import tqdm

dbname = "marketdata.db"

import io

from google.colab import files

# uploaded = files.upload()
```

```python
# Extraction des données : on part d'un csv, extrait de la base de donnée fournie par Lusis

# La cellule ressort ListeInit
# ListeInit va contenir les séquences de 8 points sur 4 heures

df = pd.read_csv("FRA40.csv", index_col="date", parse_dates=["date"])


data_sequences = df[["open", "high", "low", "close", "volume"]]


longueur = 48
echantillon = 8
nbPoints = 15000

n_in = echantillon
latent_dim = 3
input_dim = 1


openDf = df["open"].tolist()


# Liste va contenir les séquences de 8 points sur 4 heures

Liste = []
for i in range(0, len(openDf) - longueur, longueur):
    l = openDf[i : i + longueur]
    lComp = []
    for k in range(0, len(l), math.floor(len(l) / echantillon)):
        lComp.append(l[k])
    Liste.append(np.array(lComp))

# On normalise les séquences de 8 points pour pouvoir les adapter aux modèles

Liste = np.array([preprocessing.scale(l) for l in Liste])
Liste = np.reshape(Liste, (len(Liste), echantillon, input_dim))
```

```python
# Travail d'autoencodeurs utilisé par le groupe relatif
# On utilise ListInit crée précédemment (= séquences de 8 points)
# La cellule fournit Liste, qui est la version codée de ListeInit : chaque séquence temporelle de
# 8 points est codée en un vecteur à 4 coordonnées, pour réduire la dimension

inputs = Input(shape=(n_in, input_dim))
encoder = LSTM(latent_dim, activation="relu")(inputs)
decoder = RepeatVector(n_in)(encoder)
decoder = LSTM(latent_dim, activation="relu", return_sequences=True)(decoder)
decoder = TimeDistributed(Dense(input_dim))(decoder)
encoder_model = Model(inputs=inputs, outputs=[encoder])
model = Model(inputs=inputs, outputs=[decoder])

model.compile(optimizer="adam", loss="mse")
history = model.fit(
    Liste,
    Liste,
    validation_split=0.05,
    epochs=50,
    batch_size=130,
    verbose=1,
    shuffle=True,
)
decoded_liste = model.predict(Liste)
encoded_liste = encoder_model.predict(Liste)
```

```python
# fonction qui fait juste tourner kmeans pour un nombre "nbClusters" de clusters


def clusters(nbClusters):

    kmeans = KMeans(n_clusters=nbClusters)
    return kmeans
```

```python
# On définit la distance euclidienne
def d(a, b):
    return np.linalg.norm(a - b)


# Méthode qui permet d'évaluer la dispersion des données, l'efficacité de la prédiction
# pour une exécution de clustering
def variance(k, y, l):
    v = 0
    for i in range(len(l)):
        v += d(l[i], k.cluster_centers_[y[i]])
    return v
```

```python
# Méthode qui permet de comparer deux sets de clusters : pour chaque point des données,
# on ajoute la distance entre les centroids des clusters de chaque modèle auquel le point appartient
def comparer(k1, k2, y1, y2, l):
    di = 0
    for i in range(len(l)):
        di += d(k1.cluster_centers_[y1[i]], k2.cluster_centers_[y2[i]])
    return di
```

```python
# On trace la variance relative des clusters en fonction du nombre de clusters choisi
x = np.arange(100, 1400, 100)
y = []
z = []

for n in x:
    k = clusters(n)
    k2 = clusters(n)
    v = variance(k, k.fit_predict(Liste), Liste)
    z.append(comparer(k, k2, k.fit_predict(Liste), k2.fit_predict(Liste), Liste))
    y.append(v)
    print(v)

plt.xlabel("Nombre de clusters (paramètre choisi par l'utilisateur)")
plt.ylabel("Variance relative : grossièreté de la classification")
plt.plot(x, y)
# plt.plot(x,z)
```

```python
def filtrerClusters1(
    A, itA, B, seuil
):  # On va faire l'intersection de deux sets de clusters, A et B sont des tableaux de centroids, on raffine le set A
    # itA est le nombre de raffinements que A a déjà subi
    new = []
    for centroidA in A:
        for centroidB in B:
            if d(centroidA, centroidB) < seuil:
                new.append(((itA + 1) * centroidA + centroidB) / (itA + 2))
                break
    return new
```

```python
# On a les données de départ dans Liste, une marge de tolérance seuil pour la filtration.
# On fait nit fois appel à k-means, et on garde l'intersection de tous les sets créés


def nRaffinements2(nit, nbClusters, seuil, Liste):
    kmeans = clusters(nbClusters)
    y = kmeans.fit_predict(Liste)
    A = kmeans.cluster_centers_
    long = [len(A)]
    for i in range(nit):
        kmeans2 = clusters(nbClusters)
        y2 = kmeans2.fit_predict(Liste)
        B = kmeans2.cluster_centers_
        A = filtrerClusters1(A, i, B, seuil)
        long.append(len(A))

    return (A, long)
```

```python
def var(A, B, Liste):
    sA = 0
    sB = 0
    for el in Liste:
        indA = 0
        for i in range(len(A)):
            if float(np.linalg.norm(A[i] - el)) < float(np.linalg.norm(A[indA] - el)):
                indA = i

        indB = 0
        for i in range(len(B)):
            if np.linalg.norm(B[i] - el) < np.linalg.norm(B[indB] - el):
                indB = i

        sA += np.linalg.norm(A[indA] - el)
        sB += np.linalg.norm(B[indB] - el)

    return sA, sB
```

```python
# Cette fonction permet de comparer la distance de deux sets de centroids en utilisant une liste de points Liste
# Pour chaque point de Liste, on regarde pour chacune des classifications A et B, à quel cluster le point appartient,
# et on ajoute la distance entre les deux centroids à la somme totale
# Cette méthode a l'avantage, par rapport à la méthode "comparer", qu'on n'a pas besoin du kmeans total,
# seul le tableau de centroids suffit. C'est nécessaire pour la méthode "synthèse"


def comparerSets(A, B, Liste):
    somme = 0
    for el in Liste:
        indA = 0
        for i in range(len(A)):
            if float(np.linalg.norm(A[i] - el)) < float(np.linalg.norm(A[indA] - el)):
                indA = i

        indB = 0
        for i in range(len(B)):
            if np.linalg.norm(B[i] - el) < np.linalg.norm(B[indB] - el):
                indB = i

        somme += np.linalg.norm(A[indA] - B[indB])

    return somme
```

```python
# Cette méthode représente l'aboutissement de tout le travail précédent.
# Elle compare les instabilités de deux processus :
# - Le premier, en faisant tourner deux fois le kmeans classique, on


def synthese(nit, nbC, seuil, Liste):
    (set1, long) = nRaffinements2(nit, nbC, seuil, Liste)
    plt.plot(range(len(long)), long)
    plt.xlabel("Nombre de raffinements")
    plt.ylabel("Nombre de clusters restant")
    plt.show()
    print(long)

    (set2, long2) = nRaffinements2(nit, nbC, seuil, Liste)
    plt.plot(range(len(long2)), long2)
    plt.xlabel("Nombre de raffinements")
    plt.ylabel("Nombre de clusters restant")
    plt.show()
    print(long2)

    k1 = clusters(long[-1])
    k2 = clusters(long2[-1])
    y1 = k1.fit_predict(Liste)
    y2 = k2.fit_predict(Liste)
    return comparerSets(set1, set2, Liste), comparer(k1, k2, y1, y2, Liste)
```

```python
# On affiche proprement le résultat de la méthode "synthese"
def afficher(L):
    print("Résultats bruts : \n")
    print(
        "Distance entre deux sets, avec notre méthode d'intersection // Distance entre deux exécutions de l'algorithme k-means"
    )
    gain = []
    for el in L:
        gain.append(str(100 * (el[1] - el[0]) / el[1]) + "%")
    print(np.array(L))

    moyenne1 = 0
    moyenne2 = 0
    moyenneGain = 0
    for el in L:
        moyenne1 += el[0]
        moyenne2 += el[1]

    print(np.array(gain))
    moyenne1 /= len(L)
    moyenne2 /= len(L)
    print(moyenne1)
    print(moyenne2)
    print("Gain moyen :")
    print(100 * (moyenne2 - moyenne1) / moyenne2)
```

```python
# On affiche proprement le résultat de la méthode "synthese"
def sortie(L):
    fichier = open("Synthese.txt", "a")
    tx = "Résultats bruts : \nDistance entre deux sets, avec notre méthode d'intersection \nDistance entre deux exécutions de l'algorithme k-means"
    gain = []
    for el in L:
        gain.append(str(100 * (el[1] - el[0]) / el[1]) + "%")
    tx += str(np.array(L)) + "\n"

    moyenne1 = 0
    moyenne2 = 0
    moyenneGain = 0
    for el in L:
        moyenne1 += el[0]
        moyenne2 += el[1]

    tx += "Gains :\n"
    tx += str(np.array(gain)) + "\n"
    moyenne1 /= len(L)
    moyenne2 /= len(L)
    tx += "Moyenne avec notre méthode d'intersection :\n"
    tx += str(moyenne1) + "\n"
    tx += "Moyenne avec k-means classique :\n"
    tx += str(moyenne2) + "\n"
    tx += "Gain moyen :\n"
    tx += str(100 * (moyenne2 - moyenne1) / moyenne2)
    fichier.write(tx)
    fichier.close()
```

```python
# On sollicite plusieurs fois la méthode synthese, dans le but de pouvoir évaluer le gain moyen
# Chaque appel à synthese fait une comparaison entre deux méthodes, et ici, on fait plusieurs appels pour
# moyenner et obtenir une conclusion fiable

# Paramètres pour la synthèse:

# Nombre d'itérations de l'appel à synthèse pour conclure:
n = 5
nit = 30
nbClusters = 700
Li = Liste
# seuil = 0.8
s = 0.08

L = []
for _ in range(n):
    L.append(synthese(nit, nbClusters, s, Li))
```

```python
# Cette méthode, utilisée pour le backtesting, permet, à la main,
# de donner le cluster auquel un point doit être associé


def classer(ListeClusters, point):
    centroid = ListeClusters[0]
    ind = 0
    for i in range(len(ListeClusters)):
        if d(point, ListeClusters[i]) < d(centroid, ListeClusters[i]):
            ind = i
            centroid = ListeClusters[i]

    return centroid, ind
```

```python
# Cette méthode permet d'encoder un point donné ensuite, avec l'autoencodeur défini plus tôt pour l'apprentissage. Cela sert pour le backtesting


def encoderPoint(point):
    return encoder_model.predict(np.reshape(np.array(point), (1, 8, 1)))
```

```python
# Cette méthode permet, à partir d'une liste de centroids de points auto-encodés, de classer un point non auto-encodé dans la classification


def classerPointBrut(ListeClusters, pointBrut):
    return classer(ListeClusters, encoderPoint(pointBrut))
```
